```python
# read batch_job.txt
df_convert = pd.read_csv("processed/batch_job.txt", sep="\t", skiprows=2,dtype=str)[["Input", "GRCh38_CHR", "GRCh38_POS", 
                                                                           "GRCh38_REF", "GRCh38_ALT"]].fillna("")
df_convert = df_convert.drop_duplicates()
```

```python
convert_dict = {}
for index, row in df_convert.iterrows():
    pos = row["Input"]
    if row["GRCh38_CHR"] != "" and row["GRCh38_POS"] != "" and row["GRCh38_REF"] != "" and row["GRCh38_ALT"] != "":
        convert_pos = "chr{}:{}:{}>{}".format(row["GRCh38_CHR"], row["GRCh38_POS"], row["GRCh38_REF"], row["GRCh38_ALT"])
    else:
        convert_pos = ""
    convert_dict[pos] = convert_pos
```

```python
var_list = []
pos_list = []
convert_pos_list = []
for index, row in df_pos.iterrows():
    var = row["variant_name"]
    position = row["position"].split(",")
    for pos in position:
        var_list.append(var)
        pos_list.append(pos)
        convert_pos_list.append(convert_dict.get(pos, ""))
```

```python
pd.DataFrame({
    "haplotype_name": var_list,
    "position": pos_list,
    "convert_position": convert_pos_list
}).to_csv("processed/convert_position.csv", index=False)
```
