```python
translate_str_list = list(pd.read_csv("translate/en_str.csv")["en_str"].values)
```

```python
translate_en_list = []
translate_cn_list = []
count = 0
for t_str in translate_str_list:
    count += 1
    if count != 0 and count % 1000 == 0:
        pd.DataFrame({
            "en_str": translate_en_list,
            "cn_str": translate_cn_list
        }).to_csv("translate/translate_{}.csv".format(count), index=False)
        translate_en_list = []
        translate_cn_list = []
    
    try:
        result = translator.translate(t_str, src="en", dest="zh-CN")
        translate_en_list.append(t_str)
        translate_cn_list.append(result.text)
    except:
        continue

if len(translate_en_list) > 0:
    pd.DataFrame({
            "en_str": translate_en_list,
            "cn_str": translate_cn_list
        }).to_csv("translate/translate_{}.csv".format(count), index=False)
```

```python
df_translate = pd.DataFrame(columns=["en_str", "cn_str"])
for path in os.listdir("translate"):
    if "translate_" in path:
        df_translate = pd.concat(
            [
                df_translate,
                pd.read_csv(os.path.join("translate", path), dtype=str, encoding="utf-8").fillna("")
            ],
            ignore_index=True,
            axis=0
        )
df_translate.to_csv("processed/translate.csv", index=False, encoding="utf-8")
```
